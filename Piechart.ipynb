Create a PieChart

```python
from matplotlib import pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
```

wedgeprops : edgecolour represent the outpline of the diagram

```python
slices=[120,80,20,55]
labels=['Sixty','Forty','twenty','fiftyfive']
colors=['blue','orange','green','yellow']
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'},colors=colors)
plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```

using color code

```python
slices=[120,80,20,55]
labels=['Sixty','Forty','twenty','fiftyfive']
colors=['#008fd5','#fc4f30','#e5ae37','#6d904f']
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'},colors=colors)
plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```

import real time data

```python
from matplotlib import pyplot as plt
popularity=[78945,72134,57894,41234,47845]
language=['javascript','python','java','c','ruby']
colors=['orange','blue','yellow','pink','purple']
explode=[0,0,0.5,0,0]
plt.pie(popularity, labels=language, wedgeprops={'edgecolor':'black'}, explode=explode, colors=colors, startangle=90, autopct='%1.1f%%', shadow=True)
plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```
