```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("developerdata.csv")
df.head(10)
```

```python
age=df['Age']
dev_salaries=df['All_Devs']
py_salaries=df['Python']
js_salaries=df['JavaScript']
```

alpha is the parameter that control transparency control

```python
plt.plot(age,dev_salaries,label="All Developers salary",color='#444444',linestyle='--',linewidth=4)
plt.plot(age,py_salaries,label="Python Developers salary",color='#808088')
plt.plot(age,js_salaries,label="Js Developer salary")
plt.fill_between(age,py_salaries,color="orange",alpha=0.5)
plt.tight_layout()
plt.legend()
plt.title("Developers Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
median_all_devs = df['All_Devs'].median()
median_python = df['Python'].median()
median_javascript = df['JavaScript'].median()


print(f"Median salary for All Developers: {median_all_devs:,.2f}")
print(f"Median salary for Python Developers: {median_python:,.2f}")
print(f"Median salary for JavaScript Developers: {median_javascript:,.2f}")
```

where used to python developer salary

The interpolate=True argument is crucial for plt.fill_between when you're filling areas based on a where condition. It ensures that the filled region accurately connects to the intersection points where the condition changes. Without it, the filled area can appear blocky or incomplete, especially when the line (py_salaries in this case) crosses the reference line (overall_median). By setting interpolate=True, matplotlib calculates these precise intersection points, resulting in a smooth and correct fill that extends right up to where the lines cross

```python
plt.plot(age,dev_salaries,label="All Developers salary",color='#444444',linestyle='--',linewidth=4)
plt.plot(age,py_salaries,label="Python Developers salary",color='#808088')
overall_median=57287
plt.fill_between(age,py_salaries,overall_median,where=(py_salaries>overall_median),interpolate=True)
plt.fill_between(age,py_salaries,overall_median,where=(py_salaries<=overall_median),interpolate=True)

plt.tight_layout()
plt.legend()
plt.title("Developers Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
