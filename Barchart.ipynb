```python
df=pd.read_csv("data.csv")
df.head(10)
```

seperate the data like ids and languages

```python
ids=df['Responder_id']
languages=df['LanguagesWorkedWith']
print(ids)
print(languages)
```

Count the Number of languges are present , Counter() function are used to count the value

```python
from collections import Counter
language_counter=Counter()
```

for loop are used to taken data and store and split the data

```python
for response in languages.dropna():
  language_counter.update(response.split(';'))
```

store the language and id value in array fomat

```python
language=[]
popularity=[]
```

1--> java/python
2-->python/c/c++  so that id stores[0] and language stores [1] position

most_common(20) function are used to pick first higher 20 languages are taken

```python
language=[]
popularity=[]
for item in language_counter.most_common(20):
  language.append(item[0])
  popularity.append(item[1])

# Reverse the order for better visualization in horizontal bar chart
language.reverse()
popularity.reverse()
```

```python
plt.bar(language,popularity)
plt.xlabel("Language")
plt.ylabel("popularity")
plt.tight_layout()
plt.show()
```

swape the x and y axis and applying style and grid and change the bar in horizontal , height give some space

```python
plt.style.use('fivethirtyeight')

# Increase figure size for better label visibility
plt.figure(figsize=(10, 8))

plt.barh(language,popularity, color='orange', height=0.7)
plt.title("Popular Programming Language")
plt.xlabel("popularity")
plt.ylabel("language")
plt.tight_layout()
plt.grid(True)
plt.show()
```
