Naive Bayes Algorithm are used to perform data analysis


```python
import pandas as pd
import numpy as np
df=pd.read_csv("tested.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','Fare','Cabin','Embarked','Ticket','SibSp'],axis='columns',inplace=True)
df.head()
```

set survived is a target and drop from the column

```python
target=df.Survived
inputs=df.drop('Survived',axis='columns')
```

create a dummy to change the number format in the gender column

```python
dummies=pd.get_dummies(inputs.Sex).astype(int)
dummies.head(5)
```

These column should added to the existing "input" table

```python
inputs=pd.concat([inputs,dummies],axis='columns')
inputs.head(5)
```

if you add some column to existing column using this

inputs=pd.concat([inputs,existing column name],axis='columns')

drop out sex column "inplace=True" used for changes is applied in the orginal data file , without means just apply the changes taken copy of the file

```python
inputs.drop('Sex',axis='columns',inplace=True)
inputs.head()
```

Check if any column values are null

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

if you found where is null means use this

```python
inputs[inputs.Age.isnull()]
```

it shows age column so fill the value using "mean()"

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

To perform train and test the data

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

split data into train and test

Case 1 : WITHOUT random_state
train_test_split(data, test_size=0.2)

First time you run:

Train  A, B, C, D

Test  E

Second time you run:

Train  A, C, D, E

Test  B

Third time you run:

Train  B, C, D, E

Test  A

 Test data changes every time
 Result (accuracy) also changes
 You get confused 

```python
X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2,random_state=42)
```

check the split sizes

```python
print("Training data size : ", X_train.shape)
print("Testing data size :",X_test.shape)
```

X_train  Input data for learning (334 rows)

y_train  Correct answers for training

X_test  New data for testing (84 rows)

y_test  Correct answers for testing

Import Naive Bayes model

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

Make predictions using test data

```python
y_pred=model.predict(X_test)
```

Model predicts survival for 84 new passengers

y_pred  predicted answers

```python
print(y_pred[:10])
print(y_test.values[:10])
```

First line  Model prediction

Second line  Actual result

Check accuracy (final result)

```python
from sklearn.metrics import accuracy_score

accuracy=accuracy_score(y_test,y_pred)
print("Accuracy :",accuracy)
```
