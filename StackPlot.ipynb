```python
from matplotlib import pyplot as plt
```

creata a minutes

```python
minute=[1,2,3,4,5,6,7,8,9]
```

create a three player working time

```python
player1=[8,6,5,5,4,2,1,1,0]
player2=[0,1,2,2,2,4,4,4,4]
player3=[0,1,1,1,2,2,3,3,4]
```

```python
labels=['player1','player2','player3']
colors=['#6d904f','#fc4f30','#008fd5']
```

The stack diagram look vertically to analysis

```python
plt.stackplot(minute,player1,player2,player3,labels=labels,colors=colors)
plt.legend(loc=(0.07,0.05))
plt.title("Players working Time")
plt.xlabel("minute")
plt.ylabel("player work")
plt.tight_layout()
plt.show()
```
